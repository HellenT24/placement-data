### Importing Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://gist.githubusercontent.com/noamross/e5d3e859aa0c794be10b/raw/b999fb4425b54c63cab088c0ce2c0d6ce961a563/cars.csv'
df = pd.read_csv(url)
df.head()
```

```python
plt.scatter(df['hp'], df['mpg'], color='red', marker = '*')
plt.xlabel('HP', fontsize = 15)
plt.ylabel('MPG', fontsize = 15)
plt.show()
```

```python
sns.scatterplot(x ='hp', y ='mpg', data = df, 
            hue ='gear', size ='gear', style="gear",
            palette='deep', sizes=(50, 300))
plt.xlabel("hp", fontsize=15)
plt.ylabel("mpg", fontsize=15)
plt.title("scatter plot between HP and MPG")
plt.show()
```

```python
sns.scatterplot(x ='hp', y ='mpg', data = df, 
            hue ='carb', size ='carb', style="carb",
            palette='deep', sizes=(50, 450))
plt.xlabel("hp", fontsize=15)
plt.ylabel("mpg", fontsize=15)
plt.title("scatter plot between HP and MPG")
plt.show()
```

```python
income = ['Salary', 'school_fees', 'Rent', 'Shopping', 'Maintenance', 'food', 'gift']
expenditure = [2000, 5000, 4000, 1000, 7000, 1660, 850]

import matplotlib.pyplot as plt
plt.pie(expenditure, labels=income, radius=1.3, autopct='%1.0f%%', 
        explode = [0.5,0,0,0.2,0,0,0], shadow = True, startangle=45)
plt.savefig('piechart.png', bbox_inches='tight', pad_inches=2, transparent=True)
```
